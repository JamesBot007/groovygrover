$\textbf{Grover Search Algorithm}$<br>
What does it do?<br>
It solves the problem of unstructured search.
<br>
<br>
$\textbf{The Problem}:$ <br>
Size of database: $N=2^n$<br>
For a function $f(x)$, $x\in \{0,1,2,...,2^{n-1}\}$, <br>
we want to find x*, for which:<br>
$$f(x)=0,$$<br>
$$f(x*)=1.$$<br>
<br>
$\textbf{What exactly is $f(x)$}:?$<br>
$f$ is a blackbox function.<br>
The problem can also be written as:<br>
$$g(x*)=y,$$<br>
$$\text{Find x* given y},$$ <br>
$$f(x): \text{competes}\; y'=g(x)$$<br>
$$\text{compares } y'=y?\text{ and return } 1 \text{ if true.}$$<br>
In the form of a database search, this means:<br>
-$y$ is an entry in that database,<br>
-$x*$ is the index that produces $y$.<br>
<br>

$\textbf{Trick!}$<br>
Instead  write:
$$O\ket{x}=(-1)^{0}\ket{x}=(-1)^{f(x)}\ket{x}, x \neq x*$$<br>
$$O\ket{x*}=(-1)^{1}\ket{x*}=(-1)^{f(x)}\ket{x*}!$$<br>
<br>
$\textbf{The Algorithm}$<br>
Steps:<br>
1.) Initialize all n qubits to $\ket{0}$<br>
2.) Transform into uniform superposition <br>
3.) Query state by adding oracle $O$ gate<br>
4.) Increase amplitude of $x*$ absolutely using $\textit{Grover diffusion operator D}$<br>
5.) Repeate step 3.) and 4.) until $x*$ is found with a certain probability <br>
6.) Measure <br>

$\textbf{Amplitude Amplification}$<br>
Goal:Increase amplitude of $x*$ using $D$.
If $\alpha_x$ is the amplitude of x, we can define the average as:
$$\mu=\sqrt{1}{N}\sum_x \alpha_x.$$
Using $\mu$, we can define the mapping of the gate $D$ as:
$$\sum_x \alpha_x \ket{x} \to  \sum_x (2\mu-\alpha_x)\ket{x}$$
